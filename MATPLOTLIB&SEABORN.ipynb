```python
import matplotlib.pyplot as plt
a = [230,560,780,127,128]
b = [200,160,270,127,400]
years = [1,2,3,4]
profit_a = [(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b = [(b[i]-b[i-1]) for i in range(1,len(b))]
plt.plot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth = '3',label =
'CompanyA',marker='>',ms='15',mec='k')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='yellow',linestyle='dotted',label =
'CompanyB',marker = 'H')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
print(iris)
sns.scatterplot(x="sepal_length",y="petal_width",data=iris,label="mouni")
plt.title("Scatter plot of Sepal vs Petal")
plt.xlabel("sepal($)")
plt.ylabel("petal($)")
plt.legend(loc="best")
plt.show()
```
